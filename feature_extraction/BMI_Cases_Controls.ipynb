```python
# Get min,max,mean and create a new dataframe with these three values from the output of the function filter_df_onset
def min_max_mean(df,BMI):
    df_min = df.groupby([df["medical_record_number"]])['BMI'].min().to_frame()
    df_max = df.groupby([df["medical_record_number"]])['BMI'].max().to_frame()
    df_mean = df.groupby([df["medical_record_number"]])['BMI'].mean().to_frame()
    df_min.rename(columns={"BMI": BMI+"_min"}, inplace=True)
    df_max.rename(columns={"BMI": BMI+"_max"}, inplace=True)
    df_mean.rename(columns={"BMI": BMI+"_mean"}, inplace=True)
    result = df_min.merge(df_max, left_index=True, right_index=True)
    result = result.merge(df_mean, left_index=True, right_index=True)
    return result
```
