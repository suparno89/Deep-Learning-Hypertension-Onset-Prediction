```python
#creating min, max and mean by grouping MRNS and renaming them 
Pulse_Case_min =Pulse_Case.groupby([Pulse_Case.index])['pulse_rate'].min().to_frame()
Pulse_Case_max =Pulse_Case.groupby([Pulse_Case.index])['pulse_rate'].max().to_frame()
Pulse_Case_mean =Pulse_Case.groupby([Pulse_Case.index])['pulse_rate'].mean().to_frame()

Pulse_Case_min.rename(columns={"pulse_rate":'pulse_rate_min'}, inplace = True)
Pulse_Case_max.rename(columns={"pulse_rate":'pulse_rate_max'}, inplace = True)
Pulse_Case_mean.rename(columns={"pulse_rate":'pulse_rate_mean'}, inplace = True)
```

```python
# Merging the values together 
Pulse_cases = Pulse_Case_min.merge(Pulse_Case_max, left_index=True, right_index=True)
Pulse_Cases = Pulse_cases.merge(Pulse_Case_mean,left_index=True, right_index=True)
```
